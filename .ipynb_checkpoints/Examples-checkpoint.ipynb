```python
import FlashPy
import matplotlib.pyplot as plt
%matplotlib inline 
plt.rcParams['figure.figsize'] = (16,9)
```

```python
data = FlashPy.Data("../../University/2016/Honours/Work/flash4.0.1/objSST128/profile_tests/zero/SMT_hdf5_plt_cnt_0001")
```

```python
density = data.get_values('dens', unpack=True)
```

```python
fig, ax = plt.subplots()

im = ax.imshow(density.sum(axis=0), interpolation='nearest', cmap='gist_heat')

plt.colorbar(im)

plt.show()
```

```python
dens, radius = FlashPy.get_profile(data, 'dens')
```

```python
fig, ax = plt.subplots()

ax.plot(radius, dens)

plt.show()
```

```python
# make a new field for the total velocity
data.create_field("vel", ["velx*velx", "vely*vely", "velz*velz"])
```

```python
velocity = data.get_values("vel")**(0.5)
```

```python
help(data.create_field)
```

```python
vel, radius = FlashPy.get_profile(data, "vel", sqrt=True)
```

```python
fig, ax = plt.subplots()

ax.plot(radius, vel, 'b-o')

plt.show()
```

```python
data.create_field('xmom', ['dens*velx'])
```

```python
slice_information = FlashPy.slice(data, 'xmom')
```

```python
fig, ax = plt.subplots()

im = ax.imshow(slice_information, cmap='gist_heat', interpolation='nearest')

plt.colorbar(im)

plt.show()
```

```python
slice_information = FlashPy.slice(data, 'velx')
```

```python
fig, ax = plt.subplots()

im = ax.imshow(slice_information, cmap='gist_heat', interpolation='nearest')

plt.colorbar(im)

plt.show()
```
